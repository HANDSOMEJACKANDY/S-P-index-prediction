```python
def df_alginer(dfs):
    # get all columns
    index_seq = []
    for df in dfs:
        index_seq += list(df.columns)
    
    # concatenate and align
    tmp = pd.concat(dfs, axis=1, join="inner")
    tmp = tmp.reindex(columns=index_seq)
    
    # retrieve pds back
    for i in range(0, len(dfs)):
        dfs[i] = tmp.iloc[:, sum([len(dfs[j].columns) for j in range(0, i)])
                          :sum([len(dfs[j].columns) for j in range(0, i+1)])]
    return dfs
```

```python
def moving_average(df, window_size=5):
    """
    Do move average on window_size
    """
    out = df.copy()
    
    # do window_size moving average
    for i in range(window_size, len(df)+1):
        out.iloc[i-1, :] = df.iloc[i-window_size:i, :].mean()
    
    return out.iloc[window_size-1:, :]

def moving_normalize(df, mean_window_size=30, future_window_size=20):
    """
    Due to the fact that index is constantly growing, using a universal normalizing method is infeasible.
    The function use mean_window_size to calculate mean, and use the mean to divide future_window_size data.
    """
    out = df.copy()
    
    for i in range(mean_window_size, len(df)-future_window_size, future_window_size):
        out.iloc[i:i+future_window_size, :] = df.iloc[i:i+future_window_size, :] / df.iloc[i-mean_window_size:i, :].mean()
    
    i = i-future_window_size
    if not i == len(df)-future_window_size-1:
        out.iloc[i:, :] = df.iloc[i:, :] / df.iloc[i-mean_window_size:i, :].mean()
    
    return out.iloc[mean_window_size:, :] - 1

def augment_moving_normalize(df, mean_window_size=30, future_window_size=20):
    """
    Augment data
    """
    all_out = []
    for j in range(future_window_size):
        out = df.copy()

        for i in range(mean_window_size+j, len(df)-future_window_size, future_window_size):
            out.iloc[i:i+future_window_size, :] = df.iloc[i:i+future_window_size, :] / df.iloc[i-mean_window_size:i, :].mean()

        i = i-future_window_size
        if not i == len(df)-future_window_size-1:
            out.iloc[i:, :] = df.iloc[i:, :] / df.iloc[i-mean_window_size:i, :].mean()

        all_out.append(out.iloc[mean_window_size+j:, :])
    return pd.concat(all_out, axis=0)
```

```python
def generate_train_data(target, test_split=0.2, valid_split=0.1, input_len=20, output_len=10):
    """
    input is a sequence of length input_len
    output is a single value mean of length output_len
    """
    out = {}
    
    data = target.values
    
    data_len = len(data)-output_len
    test_split_len = int(data_len*(1-test_split-valid_split))
    valid_split_len = int(data_len*(1-valid_split))
    
    out["train_input"] = np.array([data[i-input_len:i].reshape((input_len, 1)) for i in range(input_len, test_split_len)])
    out["train_output"] = np.array([data[i:i+output_len].mean() for i in range(input_len, test_split_len)])
    out["test_input"] = np.array([data[i-input_len:i].reshape((input_len, 1)) for i in range(test_split_len, valid_split_len)])
    out["test_output"] = np.array([data[i:i+output_len].mean() for i in range(test_split_len, valid_split_len)])
    out["valid_input"] = np.array([data[i-input_len:i].reshape((input_len, 1)) for i in range(valid_split_len, data_len)])
    out["valid_output"] = np.array([data[i:i+output_len].mean() for i in range(valid_split_len, data_len)])
    
    return out
```

```python
# read the factor time data
df = pd.read_csv("^GSPC.csv", index_col=[0], parse_dates=[0]).iloc[1:, :-1]
```

```python
# normalize target
# target = augment_moving_normalize(df, 520 ,seq_len)
input_len = 65
output_len = 10
target = moving_normalize(df, mean_window_size=260, future_window_size=65)
target = target.iloc[:, -1]
```

```python
train_data = generate_train_data(target, test_split=0.2, valid_split=0.1, input_len=input_len, output_len=output_len)
```

```python
# flow for prev_label_input
seq_input = Input(shape=(input_len, 1))
x = seq_input
x = LSTM(units=64, activation="relu", kernel_regularizer=l2(l=0.01), return_sequences=False)(x)
x = BatchNormalization()(x)
x = Dense(units=64)(x)
x = BatchNormalization()(x)
x = Dropout(rate=0.2)(x)
output = Dense(units=1, activation="tanh")(x)

# compile model
Indexer = Model(inputs=[seq_input], outputs=[output])
Indexer.compile(loss="mse", optimizer=Adam(lr=0.005), metrics=["mse"])

# display model
plot_model(model=Indexer, to_file="temp.png")
Indexer.summary()
```

```python
Indexer.fit(x=train_data["train_input"], y=train_data['train_output'], 
            validation_data=[train_data["test_input"], train_data["test_output"]],
            epochs=300, batch_size=64, shuffle=True, verbose=1)
```

```python
def compare_output(predicted, correct):
    x = np.arange(0, correct.shape[0])
    plt.figure(figsize=(22, 2))
    sample_label, = plt.plot(x, correct, label="sample")
    predict_label, = plt.plot(x, predicted, label="predict")
    plt.legend([sample_label, predict_label], ["sample", "predict"])
```

```python
for which in ["train", "test", "valid"]:
    predicted = Indexer.predict_on_batch([train_data[which+'_input']])
    correct = train_data[which+'_output']
    compare_output(predicted, correct)
```
